```python

class Category:
    def __init__(self,name):
        
        self.name = name
        self.ledger = [] # Instance variable that stores our transactions(withdrawals,deposits,transfers)
        
        
    # Deposit money into the balance of our specific category
    def deposit(self,amount,description = ''):
        
        self.ledger.append({'amount': amount,'description': description})
        return 

    # Withdraw money from the current category, checks to see if we have sufficient funds    
    def withdraw(self,amount,description = ''):
        self.amount = 0-amount
        if self.check_funds(amount) == True:
            self.deposit(self.amount,description)
            return True
        
        return False

    # Get the sum of all transactions within a category
    def get_balance(self):
        
        balance = 0
        for entry in self.ledger:
            balance += entry['amount']
        return(balance)

    # Transfers money from one category to another
    def transfer(self,amount,destination_category):

        # Check to see if there are sufficient funds for transfer
        if self.check_funds(amount):
            
            self.withdraw(amount,description = f'Transfer to {destination_category.name}')
            destination_category.deposit(amount,description = f'Transfer from {self.name}')
            return True
            
        # Return False when there are insufficient funds
        return False

    # Checks to make sure we have enough funds to withdraw the desired amount
    def check_funds(self,amount):
        
        balance = self.get_balance()
        return (amount <= balance) # returns True when we have sufficient funds, False for insufficient funds

    def __str__(self):
        
        header = self.name.center(30,'*') + '\n'
        ledger=""
        for entry in self.ledger:
            transaction_description = "{:<23}".format(entry['description'])
            transaction_amount = "{:>7.2f}".format(entry['amount'])
            ledger += "{}{}\n".format(transaction_description[:23],transaction_amount[:7])
        total = "Total: {:.2f}".format(self.get_balance())
        return header + ledger + total



def create_spend_chart(categories):
        
    title = "Percentage spent by category\n"
    lines = {}
    x_axis = '    -'
    total_spent = 0
    spending_by_category = {}

     
    for i in list(range(100,0,-10)):
        if i < 100:
            lines[i] = f' {i}|'
        else:
            lines[i] = f'{i}|'
    
    # Calculate the total amount spent
    for category in categories:
        x_axis += '---'
        for item in category.ledger:
            if item['amount'] < 0:
                total_spent += abs(item['amount'])
                
                
    for category in categories:
        category_spent = 0
        for item in category.ledger:
            if item['amount'] < 0:
                category_spent += abs(item['amount'])
                
        # Calculate the percentage spent by each category and round down to the nearest 10
        spending_by_category[category.name] = ((category_spent/total_spent)*100 // 10)*10
        
    for percent in spending_by_category.values():
        for key in lines.keys():
            if percent >= key:
                lines[key] += " o "
            else:
                lines[key] += "   "

    # making sure to match the exact number of spaces in output (we are going to use this as the max length for each line in output)
    #x_axis += ' '
    lines['x_axis'] = x_axis
    category_names = [category.name for category in categories]

    # hard to explain what I'm doing here honestly 
    for i in range(len(max(category_names,key=len))):
        
        lines[f'line_{i}'] = '     '
        for j in range(len(x_axis)):

            if len(lines[f'line_{i}']) < len(x_axis):
                try:
                    lines[f'line_{i}'] += category_names[j][i]
                    lines[f'line_{i}'] += '  '
                except:
                    lines[f'line_{i}'] += '   '
                    continue

    # putting together our output graph and stripping off the final \n
    output_graph = ''
    for key in lines:
        if len(lines[key]) < len(x_axis):
            lines[key] += ' '
        output_graph += lines[key] +'\n'
    output_graph = output_graph.strip('\n')
    return title + output_graph
        
        
```

```python
food = Category('Food')
clothing = Category('Clothing')
auto = Category('Auto')
entertainment = Category('Entertainment')

food.deposit(1000,'initial deposit')
clothing.deposit(100,'initial deposit')
auto.deposit(300,'initial deposit')
entertainment.deposit(200,'intial deposit')

food.withdraw(10.15,'groceries')
food.withdraw(15.89,'restaurant and more food for dessert')
clothing.withdraw(25,'t-shirt')
clothing.withdraw(12,'socks')
auto.withdraw(25,'gas')
entertainment.withdraw(12,'movie')
entertainment.withdraw(20,'Netflix')
```

```python
# shows that our balance sheet for each category is working
categories = [food,clothing,auto,entertainment]
for entry in categories:
    print(entry)
    print('\n')
```
